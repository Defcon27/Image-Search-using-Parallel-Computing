```python
import numpy as np
def chi2_distance(histA, histB, eps = 1e-10):
    # compute the chi-squared distance
    d = 0.5 * np.sum([((a - b) ** 2) / (a + b + eps) for (a, b) in zip(histA, histB)])
        
    return d
```

```python
print(1-chi2_distance([2,1,3],[3,1,2]))
```
