```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preprocessing

### Resizing

```python
image_path="Image_Database/113000.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_reduced = cv2.resize(image, (150, 150))

fig, ax = plt.subplots(1,2, figsize=(20,7))
ax[0].set_title("Origin Image")
ax[0].imshow(image)
ax[1].set_title("Resized Image")
ax[1].imshow(image_reduced)
plt.show()
```

### GreyScaling

```python
image_path="Image_Database/113000.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(1,2, figsize=(14,7))
ax[0].set_title("Origin Image")
ax[0].imshow(image)
ax[1].set_title("Grey Scaled Image")
ax[1].imshow(gray,cmap="gray")
plt.show()
```

## Noise Reduction (Low Pass Filter)

```python
from NoiseReduction import NoiseReduction

image_path="Image_Database/113000.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

low_filters = NoiseReduction(image)


fig, ax = plt.subplots(3,2, figsize=(17,17))
ax[0,0].set_title("Origin Image")
ax[0,0].imshow(low_filters.image, cmap="gray")
ax[0,1].set_title("Small Blur")
ax[0,1].imshow(low_filters.applySmallBlur(), cmap="gray")

ax[1,0].set_title("Origin Image")
ax[1,0].imshow(low_filters.image, cmap="gray")
ax[1,1].set_title("Large Blur")
ax[1,1].imshow(low_filters.applyLargeBlur(), cmap="gray")

ax[2,0].set_title("Origin Image")
ax[2,0].imshow(low_filters.image, cmap="gray")
ax[2,1].set_title("Gaussian Blur")
ax[2,1].imshow(low_filters.applyGaussianBlur(), cmap="gray")

plt.show()
```

## Convolutional Filters (High pass Filters)

### High pass Filters

```python
from ConvolutionalFilters import ConvolutionFilter

image_path="Image_Database/113000.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

high_filters = ConvolutionFilter(image)


fig, ax = plt.subplots(1,3, figsize=(14,12))
ax[0].set_title("SobelX Filter")
ax[0].imshow(high_filters.sobelX,cmap="gray")
ax[1].set_title("SobelY Filter")
ax[1].imshow(high_filters.sobelY, cmap="gray")
ax[2].set_title("Laplacian")
ax[2].imshow(high_filters.laplacian, cmap="gray")
plt.show()
```

### Convolving Image with Filters

```python
from ConvolutionalFilters import ConvolutionFilter

image_path="Image_Database/113000.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

high_filters = ConvolutionFilter(image)


fig, ax = plt.subplots(4,2, figsize=(17,17))
ax[0,0].set_title("Origin Image")
ax[0,0].imshow(high_filters.image, cmap="gray")
ax[0,1].set_title("Sobel X")
ax[0,1].imshow(high_filters.applySobelX(), cmap="gray")

ax[1,0].set_title("Origin Image")
ax[1,0].imshow(high_filters.image, cmap="gray")
ax[1,1].set_title("Sobel Y")
ax[1,1].imshow(high_filters.applySobelY(), cmap="gray")

ax[2,0].set_title("Origin Image")
ax[2,0].imshow(high_filters.image, cmap="gray")
ax[2,1].set_title("Laplacian")
ax[2,1].imshow(high_filters.applyLaplacian(), cmap="gray")

ax[3,0].set_title("Origin Image")
ax[3,0].imshow(high_filters.image, cmap="gray")
ax[3,1].set_title("Canny Edge")
ax[3,1].imshow(high_filters.applyCannyEdge(), cmap="gray")

plt.show()
```
